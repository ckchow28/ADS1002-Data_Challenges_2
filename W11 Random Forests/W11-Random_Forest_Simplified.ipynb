### What is Random Forest in Machine Learning?

Random Forest is an ensemble learning method used for classification and regression tasks. In simple terms, it is like a "forest" of many "decision trees" working together to make a prediction. The idea is that by combining the results of multiple decision trees, you can get a more accurate and stable prediction than using just one tree.

### How Random Forest Works

1. **Creating Multiple Decision Trees**:
- The Random Forest algorithm starts by creating multiple decision trees. Each tree is trained on a different subset of the training data, chosen randomly with replacement (this is known as **bootstrap sampling**). This means that some data points may be used more than once in the training of a single tree, while others may not be used at all.
2. **Random Feature Selection**:
- When growing each tree, the algorithm selects a random subset of features (input variables) at each split point, instead of considering all available features. This randomness helps in making the trees less correlated, improving the overall prediction when they are combined.
3. **Making Predictions**:
- For classification tasks, when you want to make a prediction on new data, each tree in the forest "votes" for a class label. The class with the most votes across all trees is chosen as the final prediction.
- For regression tasks, the predictions from all the trees are averaged to get the final result.

### How is Random Forest Different from a Decision Tree?

- **Single Tree vs. Multiple Trees**:
    - A decision tree is a single model that splits the data based on feature values to make predictions. In contrast, Random Forest uses multiple decision trees, each trained on different data and features, and combines their predictions.

- **Overfitting**:
    - A single decision tree can easily overfit the training data, especially if it's very deep (has many levels). Overfitting means it performs well on the training data but poorly on new, unseen data. Random Forest reduces the risk of overfitting because it averages the predictions of many trees, which tends to smooth out errors that any individual tree might make.

- **Stability**:
    - Decision trees are sensitive to small changes in the data. A small change can lead to a completely different tree. Random Forest, by aggregating many trees, provides more stable and reliable predictions.

### What is Bagging and Pasting in Random Forest?

In Random Forest, the key idea is to create multiple decision trees and then combine their predictions. To generate these multiple trees, two techniques called **Bagging** and **Pasting** are commonly used.

#### What is Bagging?

**Bagging** stands for **Bootstrap Aggregating**. It is a technique used to reduce the variance of a model by training multiple models (like decision trees) on different subsets of the training data and then averaging or voting on their predictions.

#### How Bagging Works:

1. **Data Sampling with Replacement**:
    - In bagging, you create multiple subsets of your training data by sampling with replacement. This means that each time you pick an example from your original dataset, you put it back before picking the next one. As a result, some examples might appear multiple times in a subset, while others might not appear at all.
2. **Train Multiple Models**:
    - You then train a separate decision tree on each of these subsets.
3. **Aggregate the Results**:
    - Finally, when making a prediction, you take the average (for regression) or the majority vote (for classification) of the predictions made by all the trees.

**Simple Example of Bagging**:

Imagine you have a dataset with 10,000 examples. Using bagging, you might create 100 different subsets, each containing 10,000 examples (some repeated, some missing). You train 100 decision trees, each on one of these subsets. When predicting, you combine the results from all 100 trees.

**Sampling with Replacement**

- In sampling with replacement, after selecting an item, you **put it back** into the dataset before selecting the next one. This means that the same item can be picked **more than once**.
- **Example**: Imagine you have a bowl of 5 numbered balls: {1, 2, 3, 4, 5}. If you're picking a ball, noting the number, and then **putting it back in the bowl** before picking again, you might pick the same ball multiple times.
- **Real-world Example**: A raffle draw where after each person wins a prize, their name is put back into the raffle for another chance to win again. The same person could win multiple prizes because their name is returned to the pool.

**Sampling without Replacement**:

- In **sampling without replacement**, after selecting an item, you **do not put it back** into the dataset. Once an item is picked, it cannot be selected again.
- **Example**: Using the same bowl of 5 numbered balls: {1, 2, 3, 4, 5}. This time, when you pick a ball, you don’t put it back. Each ball is only selected once, and every new pick reduces the number of balls left in the bowl.
- **Real-world Example**: A lottery where once a person wins a prize, they are not eligible to win again. After they win, their name is removed from the pool.

**Key Differences**:

- **With Replacement**: Items (or data points) can be selected more than once.
- **Without Replacement**: Once an item is selected, it cannot be chosen again.

Both methods are important in machine learning, with sampling without replacement more commonly used in real-world scenarios like surveys (where you don’t ask the same person twice). Sampling with replacement is more useful in simulations or certain types of experiments where repetition of events is needed.

#### What is Pasting?

**Pasting** is similar to bagging, but with one key difference: it involves sampling without replacement.

#### How Pasting Works:

1. **Data Sampling without Replacement**:
    - In pasting, you also create multiple subsets of your training data, but this time without replacement. Each example in your original dataset can only appear once in each subset.
2. **Train Multiple Models**:
    - Like bagging, you train a separate decision tree on each of these subsets.
3. **Aggregate the Results**:
    - Finally, you combine the predictions from all the trees, just like in bagging.

**Simple Example of Pasting**:

Using the same dataset with 10,000 examples, you might create 100 different subsets, each containing 5,000 unique examples. You train 100 decision trees, each on one of these subsets. When predicting, you combine the results from all 100 trees.

### Role of Bagging and Pasting in Random Forest

In a Random Forest, **Bagging** is the technique that is most commonly used. It helps in:

- **Reducing Variance**: By training each tree on a different subset of the data, Random Forest reduces the model's variance, leading to more stable and accurate predictions.
- **Handling Overfitting**: Bagging helps prevent overfitting because it ensures that the individual trees are not too similar, which would happen if they were trained on the same data.

**Pasting**, while similar, is less commonly used but can still be useful in scenarios where you want to ensure that each subset of data is unique and not repeated.

#### In Summary (Bagging, Pasting)

- **Bagging** uses data sampling with replacement to create different subsets of the training data, trains multiple models on these subsets, and aggregates their predictions to reduce variance and improve accuracy.
- **Pasting** uses data sampling without replacement, ensuring that subsets are unique, and also aggregates the predictions of multiple models.

Both techniques are essential in Random Forest to create diverse decision trees, which together lead to better overall performance.

### Is Random Forest Better than a Decision Tree?

In most cases, **Random Forest** is considered better than a single **Decision Tree**. However, whether it’s truly "better" depends on the specific situation, including the type of problem you're solving, the size of your dataset, and the need for model interpretability.

#### Why Random Forest is Generally Better:

1. **Reduces Overfitting**:
    - A single decision tree is prone to overfitting, especially if it's deep (i.e., it has many levels). Overfitting means the model becomes too tailored to the training data and fails to generalize well to new data. Random Forest, by combining the predictions of multiple trees, averages out the noise and reduces the risk of overfitting.
2. **Improves Accuracy**:
    - By aggregating the results of many decision trees, Random Forest typically produces more accurate and stable predictions than a single decision tree. Each tree might have different weaknesses, but when combined, these weaknesses tend to cancel out.
3. **Handles Variability**:
    - Decision trees are sensitive to slight changes in the data. A small change in the data can result in a completely different tree structure. Random Forest mitigates this sensitivity by using multiple trees, each trained on a different subset of the data, making the overall model more robust.

#### Simple Example to Illustrate the Difference:

Imagine you are trying to predict whether it will rain tomorrow based on historical weather data (e.g., temperature, humidity, wind speed, etc.).

**Using a Single Decision Tree**:

- You train one decision tree on your data. This tree might say:
    - If humidity is above 80%, predict "Rain".
    - If humidity is below 80% and wind speed is above 15 mph, predict "No Rain".
    This tree could work well on your training data, but it might be too simplistic or too specific, leading to poor performance on new, unseen data. If the humidity is just slightly below 80% and wind speed is 14 mph, the tree might give an incorrect prediction.

**Using a Random Forest**:

- Now, you create a Random Forest with 100 trees. Each tree might have different rules:
    - Tree 1 might say: "If humidity is above 75%, predict 'Rain'."
    - Tree 2 might say: "If temperature is below 60°F and humidity is above 85%, predict 'Rain'."
    - And so on...

    Each tree makes its prediction, and then the Random Forest takes the majority vote or averages the results to make the final prediction.

**Outcome**:

- The Random Forest prediction is less likely to be wrong because it’s based on a collective decision of many trees. It’s less sensitive to small changes in the data and less likely to overfit.

### When a Decision Tree Might Be Preferred:

- **Interpretability**: Decision trees are easy to interpret and visualize. If you need a model that you can easily explain to non-technical stakeholders, a decision tree might be better.
- **Speed**: Decision trees are faster to train and make predictions. In situations where you need quick results and have limited computational resources, a single decision tree might be sufficient.
- **Small Datasets**: For very small datasets, a decision tree might perform just as well as a Random Forest without the added complexity.

### In Summary:

- **Random Forest** is usually better than a single decision tree because it reduces overfitting, improves accuracy, and handles variability better.
- **Decision Trees** might be preferred for their simplicity, interpretability, and speed, especially in smaller datasets or when the need for model transparency is high.

### Applying Random Forest algorithm in Python 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from matplotlib.colors import ListedColormap
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
wine_data = pd.read_csv("../data/wine_data.csv", header = None,
                 names = ['Class','Alcohol','MalicAcid','Ash','Alcalinity','Mg',
                          'Phenols','Flavanoids','Nonflavanoids','Proanthocyanins',
                          'Colour','Hue','ProteinConcentration','Proline'
                         ]
                )
# Display the first few rows
wine_data.head()
```

```python
# Check for missing values
missing_values = wine_data.isnull().sum()
missing_values
```

```python
# Distribution of target classes
sns.countplot(x='Class', data=wine_data)
plt.title("Distribution of Wine Classes")
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = wine_data.corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Correlation with the target label
correlation_with_target = correlation_matrix["Class"].drop("Class").sort_values(ascending=False)
print(correlation_with_target)
```

```python
# Selecting features with high correlation to the target label
selected_features = correlation_with_target.index[correlation_with_target.abs() > 0.2].tolist()
print("Selected features:", selected_features)

# Creating the feature matrix X and target vector y
X = wine_data[selected_features]
y = wine_data["Class"]
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

# Evaluate the model
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```

### How Random Forest Determines Feature Importance:

1. **Tree Construction**: Each decision tree in the random forest selects features based on the criteria that best splits the data to reduce impurity (like Gini impurity or entropy for classification tasks).
2. **Node Splits**: During training, each node in the trees is split based on one of the features. The feature that gives the best split (reduces impurity the most) is chosen.
3. **Feature Contribution**: Random Forest measures how much impurity (e.g., Gini impurity) is reduced by each feature across all trees. Features that reduce more impurity are considered more important.
4. **Averaging Across Trees**: Random Forest aggregates these importance scores across all the trees in the ensemble to calculate a final importance score for each feature.

The higher the score, the more important that feature is for making predictions in the model.

```python
# Feature importance plot

feature_importances = rf_model.feature_importances_
indices = np.argsort(feature_importances)[::-1]

# Create a DataFrame for easier plotting
features_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})

# Sort features by importance
features_df = features_df.sort_values(by='Importance', ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=features_df, palette='viridis')
plt.title("Feature Importance in Random Forest Classifier (Wine Dataset)")
plt.show()
```

```python
# Train a Decision Tree to demonstrate overfitting

# Initialize and train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Make predictions
y_train_pred_dt = dt_model.predict(X_train)
y_test_pred_dt = dt_model.predict(X_test)

# Evaluate the model on training data (this will likely be very high due to overfitting)
train_accuracy_dt = accuracy_score(y_train, y_train_pred_dt)
test_accuracy_dt = accuracy_score(y_test, y_test_pred_dt)
print(f"Decision Tree Training Accuracy: {train_accuracy_dt:.2f}" )
print(f"Decision Tree Test Accuracy: {test_accuracy_dt:.2f}")
```

**Note**: If the training accuracy is much higher than the test accuracy, it indicates that the model is overfitting.

In the above code, DT classifier training accuracy is 100% and test accuracy is 96%. This means model is overfitting!.

```python
# Train a Random Forest to reduce overfitting
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
y_train_pred_rf = rf_model.predict(X_train)
y_test_pred_rf = rf_model.predict(X_test)
train_accuracy_rf = accuracy_score(y_train, y_train_pred_rf)
test_accuracy_rf = accuracy_score(y_test, y_test_pred_rf)

# Evaluate the model on both training and test data
print(f"Random Forest Training Accuracy: {train_accuracy_rf:.2f}" )
print(f"Random Forest Test Accuracy: {test_accuracy_rf:.2f}")
```

**Note**: In Random Forest, the training accuracy is still high, but the test accuracy should be much closer to the training accuracy compared to the Decision Tree, indicating reduced overfitting.

```python
# Step 5: Plot Decision Tree vs Random Forest Accuracy
model_names = ['Decision Tree', 'Random Forest']
train_accuracies = [train_accuracy_dt, train_accuracy_rf]
test_accuracies = [test_accuracy_dt, test_accuracy_rf]

plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(model_names))

# Plot training accuracy
plt.bar(index, train_accuracies, bar_width, label='Training Accuracy')

# Plot test accuracy
plt.bar([i + bar_width for i in index], test_accuracies, bar_width, label='Test Accuracy')

plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Decision Tree vs Random Forest: Overfitting Comparison')
plt.xticks([i + bar_width / 2 for i in index], model_names)
plt.legend()
plt.show()
```

#### Summarizing the Results

1. **Decision Tree**:

    - **Training Accuracy**: Is 1.00 (100%), as the model memorizes the training data.
    - **Test Accuracy**: Lower (e.g., 0.96), indicating overfitting since the model struggles to generalize to unseen data.

2. **Random Forest**:

    - **Training Accuracy**: High (e.g., around 1.00), but not as high as the Decision Tree.
    - **Test Accuracy**: Close to the training accuracy (e.g., 0.98), indicating better generalization and less overfitting.

#### Conclusion:

- **Decision Tree**: Overfits the data by memorizing the training examples, leading to a large gap between training and test accuracy.
- **Random Forest**: Reduces overfitting by averaging multiple decision trees' predictions, resulting in better generalization and closer training-test performance.
